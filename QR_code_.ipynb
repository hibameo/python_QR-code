```python
import qrcode
from PIL import Image

# Input data (change this to your link or text)
data = "https://www.google.com"  # Replace with any URL or text

# Create QR code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(data)
qr.make(fit=True)

# Create QR code image
img = qr.make_image(fill="black", back_color="white")

# Display QR code in Google Colab
img.show()
img
```
